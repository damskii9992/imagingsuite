```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xx = np.linspace(0,1023,512)
yy = np.linspace(0,1,100)
x,y = np.meshgrid(xx,yy/180*np.pi)
```

```python
dev=np.abs(x)*np.sin(y)

fig,ax = plt.subplots(1,figsize=(10,8))
#plt.imshow(dev*(dev<3),origin='lower',extent=[0,1023,0,yy.max()],aspect='auto')
th = 1
dev[2*th<dev]=2*th

ax.imshow(dev*(dev<3),origin='lower',aspect='auto',extent=(0,xx[-1],0,yy[-1]),cmap='coolwarm')

manual_locations = [
    (100, 0.125), (150, 0.25), (175, 0.375),(250,0.5)]
CS=ax.contour(dev, levels=[0.25,0.5,1,1.5],cmap='binary',extent=(0,xx[-1],0,yy[-1]))
ax.clabel(CS, inline=True, fontsize=10,manual=manual_locations)
ax.set(title='Max line cross talk with MuhRec tilt correction',
      xlabel='Sample radius in pixels',
      ylabel='Corrected tilt angle')
ax.grid('on')
fig.tight_layout()
plt.savefig('rowdeviation.svg')
plt.savefig('rowdeviation.png',dpi=300)
```
